#####Let $W_n$, $n=0,1,2,3,...,$ be a sequence of IID zero-mean Gaussian random variables with variance $Ïƒ^2_W=1$
#####(a) Write a MATLAB program to generate the process
\begin{align}
        X[n]=\frac{1}{2}X[n-1]-\frac{1}{4}X[n-2]-\frac{1}{4}X[n-3]+W_n,
    \end{align}
#####where $X[0]=W_0$, and $X[n]=0$ for $n<0$.
#####(b)Estimate the mean function of this process by generating a large number of realizations of the random process and computing the sample mean.
#####(c) Compute the time-averaged mean of the process from a single realization. Does this seem to give the same result as the ensemble mean estimated in part (b)?

```python
import numpy as np
import matplotlib.pyplot as plt

# Part (a)
sequence_length = 1000
realization_no = 1000
X_seq_realization_mat = np.zeros((realization_no, sequence_length))

for i in range(realization_no):
    x = np.zeros(sequence_length)
    for n in range(sequence_length):
        W_n = np.random.normal(0, 1)
        if n == 0:
            x[n] = W_n
        elif n == 1:
            x[n] = 0.5 * x[n - 1] + W_n
        elif n == 2:
            x[n] = 0.5 * x[n - 1] - 0.25 * x[n - 2] + W_n
        else:
            x[n] = 0.5 * x[n - 1] - 0.25 * x[n - 2] - 0.25 * x[n - 3] + W_n
    X_seq_realization_mat[i, :] = x

# Part (b)
ensemble_mean = np.mean(X_seq_realization_mat, axis=0)
plt.figure()
plt.plot(range(1, sequence_length + 1), ensemble_mean)
plt.xlabel('Realization')
plt.ylabel('Ensemble mean')
```

```python
# Part (c)
temporal_mean = np.mean(X_seq_realization_mat[0, :])
print(temporal_mean)
```

**Both the ensemble and temporal means are near zero. While the ensemble mean exhibits some variation, this variation centers around zero. Therefore, both means are in close proximity to zero.**
