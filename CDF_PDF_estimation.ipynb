#####Suppose $X_i$ are all independent and identically uniformly distributed over $[0,1], i=1,...,n$.

1. Estimate the CDF using n IID random variables for $n = 10$, $n = 100$, and $n = 1000$ and plot the estimated CDFs.

2. Estimate the PDF using n IID random variables for $n = 100$, $n = 1000$, and $n=10000$ using histogram and plot the estimated PDFs.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)  # Set a seed for reproducibility

N = [10, 100, 1000]  # Set number of samples
z = np.arange(-0.5, 1.51, 0.01)  # Define variable for horizontal axis

# Initialize the figure for CDF
plt.figure(1)

for i in N:
    x = np.random.rand(i)  # Generate uniform random samples
    F = np.zeros(len(z))  # Initialize CDF estimate
    for j in range(len(z)):
        for n in range(i):
            if x[n] < z[j]:
                F[j] += 1
    F /= i
    plt.plot(z, F, label=f'N = {i}')  # Plot results

plt.xlabel('x')
plt.ylabel('CDF $F_X(x)$')
plt.legend()

N = [100, 1000, 10000]  # Set number of samples
z = np.arange(-0.5, 1.51, 0.01)  # Define variable for horizontal axis
eps = 0.0099

# Initialize the figure for PDF
plt.figure(2)

for i in N:
    x = np.random.rand(i)  # Generate uniform random samples
    F = np.zeros(len(z))  # Initialize PDF estimate
    for j in range(len(z)):
        for n in range(i):
            if z[j] < x[n] < (z[j] + eps):
                F[j] += 1
    F /= i * eps
    plt.plot(z, F, label=f'N = {i}')  # Plot results

plt.xlabel('x')
plt.ylabel('PDF $f_X(x)$')
plt.legend()

plt.show()
```
